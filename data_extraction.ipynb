```python
# edit dataframe
nuclear_df = nuclear_df[nuclear_df['index'] == 'Total World']
solar_df = solar_df[solar_df['index'] == 'Total World']
wind_df = wind_df[wind_df['index'] == 'Total World']
coal_df = coal_df[coal_df['index'] == 'Total World']


# concatenate dataframes
df = pd.concat([nuclear_df, solar_df, wind_df, coal_df], axis=0)
df = df.drop(columns=['index'])
df = df.reset_index(drop=True)

#add a column "energy" with the type of energy
df['energy'] = ['nuclear', 'solar', 'wind', 'coal']

#delete column 24, 25 and 26
df = df.drop(columns=['2022.1', '2012-22', '2022.2'])


```
